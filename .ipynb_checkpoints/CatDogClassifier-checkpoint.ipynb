```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# --- Configuration for your dataset and model ---
# Image dimensions
IMG_WIDTH = 150
IMG_HEIGHT = 150

# Batch size for training
# You establish a batch size instead of passing all the files at once. 
BATCH_SIZE = 32

# Number of epochs for training
EPOCHS = 15 #this is one passing of the data or batch through the algorithm

# Paths to your organized data
# This assumes 'cats_vs_dogs_data' is in the same directory as your notebook
base_data_dir = 'cats_vs_dogs_data'

train_dir = os.path.join(base_data_dir, 'train') # This creates a string from the base directory to the train directory
validation_dir = os.path.join(base_data_dir, 'validation') # This can aslo be used with multiple parameters to access grandchildren folders
```

```python
print(f"Training directory: {train_dir}")
print(f"Validation directory: {validation_dir}")

if not os.path.exists(train_dir) or not os.path.exists(validation_dir):
    print("ERROR: Training or validation directory not found. Please check your 'base_data_dir' path.")
else:
    print("Train and validation directories found successfully.")
    # Let's check if the subdirectories and some images exist
    try:
        print(f"Number of training cat images: {len(os.listdir(os.path.join(train_dir, 'cats')))}")
        print(f"Number of training dog images: {len(os.listdir(os.path.join(train_dir, 'dogs')))}")
        print(f"Number of validation cat images: {len(os.listdir(os.path.join(validation_dir, 'cats')))}")
        print(f"Number of validation dog images: {len(os.listdir(os.path.join(validation_dir, 'dogs')))}")
    except FileNotFoundError:
        print("ERROR: 'cats' or 'dogs' subdirectories not found within train/validation folders. Make sure your data is organized correctly.")
```

```python
# Create training dataset
train_dataset = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    labels='inferred',
    label_mode='binary', # For cat vs dog (0 or 1)
    image_size=(IMG_HEIGHT, IMG_WIDTH),
    interpolation='nearest',
    batch_size=BATCH_SIZE,
    shuffle=True
)

# Create validation dataset
validation_dataset = tf.keras.utils.image_dataset_from_directory(
    validation_dir,
    labels='inferred',
    label_mode='binary',
    image_size=(IMG_HEIGHT, IMG_WIDTH),
    interpolation='nearest',
    batch_size=BATCH_SIZE,
    shuffle=False # No need to shuffle validation data
)
```

```python
class_names = train_dataset.class_names
print(f"Class names TensorFlow inferred: {class_names}")
# This will likely be ['cats', 'dogs'] or ['dogs', 'cats'].
# The first name corresponds to label 0, the second to label 1.
```

```python
plt.figure(figsize=(10, 10))
for images, labels in train_dataset.take(1):  # Takes one batch from the dataset
    for i in range(9): # Display the first 9 images in that batch
        if i < len(images): # Make sure we don't go out of bounds if batch is smaller than 9
            ax = plt.subplot(3, 3, i + 1)
            plt.imshow(images[i].numpy().astype("uint8")) # Convert to NumPy array to display
            plt.title(f"Label: {class_names[int(labels[i])]}")
            plt.axis("off")
plt.show() # This ensures the plot is displayed in the notebook
```
